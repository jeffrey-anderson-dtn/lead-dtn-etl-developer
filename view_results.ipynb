# View ETL Results
Use this notebook to inspect input data and output parquet files.

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("ViewResults") \
    .master("local[*]") \
    .getOrCreate()
```

## Input Data

```python
# Crop yield data
yield_df = spark.read.parquet("data/crop_yield")
print(f"Crop Yield Records: {yield_df.count()}")
yield_df.show(10)
```

```python
# Abandonment data
abandon_df = spark.read.parquet("data/county_crop_abandonment")
print(f"Abandonment Records: {abandon_df.count()}")
abandon_df.show(10)
```

## Output Data (after running solution)

```python
# Field-level production
try:
    field_df = spark.read.parquet("output/field_level_production")
    print(f"Field-Level Records: {field_df.count()}")
    field_df.show(10)
except:
    print("No field-level output yet. Run the ETL pipeline first.")
```

```python
# County rollup
try:
    county_df = spark.read.parquet("output/county_rollup")
    print(f"County Rollup Records: {county_df.count()}")
    county_df.show(10)
except:
    print("No county rollup output yet. Run the ETL pipeline first.")
```

```python
spark.stop()
```
